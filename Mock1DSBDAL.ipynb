```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
features_df = df.drop(columns = ['Species'])
```

```python
features_df.describe()
```

```python
encode = preprocessing.OneHotEncoder(sparse=False)
```

```python
encoded = encode.fit_transform(df[['Species']])
```

```python
enc_df = pd.DataFrame(encoded, columns=encode.get_feature_names_out(['Species']))
```

```python
df_encode = features_df.join(enc_df)
```
